```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cppo_mean = pd.read_csv("./logs/run-12072020183315-tag-causality_mean_causal_factor.csv")
cpg_mean = pd.read_csv(".logs/run-12072020190532-tag-causality_mean_causal_factor.csv")
cppo_e20 = pd.read_csv(".logs/run-12072020183315-tag-causality_E20.csv")
cpg_e20 = pd.read_csv(".logs/run-12072020190532-tag-causality_E20.csv")
```

```python
plt.figure()
plt.plot(cppo_mean["Value"], label="CausalPPO", c="g")
plt.title("Mean Causal Factor for Causal PPO")
plt.xlabel("Epochs")
plt.ylabel("Causal Factor")
plt.legend()
plt.savefig("plots/cppo_mean_causal_factor.png")

plt.figure()
plt.plot(cpg_mean["Value"], label="CausalPG", c="b")
plt.title("Mean Causal Factor for Causal Policy Gradient")
plt.xlabel("Epochs")
plt.ylabel("Causal Factor")
plt.legend()
plt.savefig("plots/cpg_mean_causal_factor.png")

plt.figure()
plt.plot(cppo_mean["Value"], label="CausalPPO", c="g")
plt.plot(cpg_mean["Value"], label="CausalPG", c="b")
plt.title("Mean Causal Factor")
plt.xlabel("Epochs")
plt.ylabel("Causal Factor")
plt.legend()
plt.savefig("plots/both_mean_causal_factor.png")
```

```python
plt.figure()
plt.plot(cppo_e20["Value"], label="CausalPPO", c="g")
plt.title("Episodic Causal Factor for Causal PPO")
plt.xlabel("Timestep")
plt.ylabel("Causal Factor")
plt.legend()
plt.savefig("plots/cppo_episodic_causal_factor.png")

plt.figure()
plt.plot(cpg_e20["Value"], label="CausalPG", c="b")
plt.title("Episodic Causal Factor for Causal Policy Gradient")
plt.xlabel("Timestep")
plt.ylabel("Causal Factor")
plt.legend()
plt.savefig("plots/cpg_episodic_causal_factor.png")

plt.figure()
plt.plot(cppo_e20["Value"], label="CausalPPO", c="g")
plt.plot(cpg_e20["Value"], label="CausalPG", c="b")
plt.title("Episodic Causal Factor")
plt.xlabel("Timestep")
plt.ylabel("Causal Factor")
plt.legend()
plt.savefig("plots/both_episodic_causal_factor.png")
```
